```python
#Importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the good old Iris dataset
iris = sns.load_dataset("iris")

#Construct the plot from the dataset
sns.swarmplot(x="species",y="petal_length",data=iris)

#Display the plot
plt.show()
```

```python
# Time to do something similar with the seconnd of the "Three Great Toy Datasets", the Titanic dataset.
#Load data

titanic = sns.load_dataset("titanic")

# Set up a factorplot
g = sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", palette="muted", legend=False)
                   
# Show plot
plt.show()
```

```python
#Lets try something different with the colour palette
import numpy as np
from matplotlib.colors import ListedColormap

# Define a variable N
N = 500

# Construct the colormap
current_palette = sns.color_palette("muted", n_colors=5)
cmap = ListedColormap(sns.color_palette(current_palette).as_hex())

# Initialize the data
data1 = np.random.randn(N)
data2 = np.random.randn(N)
# Assume that there are 5 possible labels
colors = np.random.randint(0,5,N)

# Create a scatter plot
plt.scatter(data1, data2, c=colors, cmap=cmap)

# Add a color bar
plt.colorbar()

# Show the plot
plt.show()
```
