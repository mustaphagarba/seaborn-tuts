```python
#Importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the good old Iris dataset
iris = sns.load_dataset("iris")

#Construct the plot from the dataset
sns.swarmplot(x="species",y="petal_length",data=iris)

#Display the plot
plt.show()
```

```python
# Time to do something similar with the seconnd of the "Three Great Toy Datasets", the Titanic dataset.
#Load data

titanic = sns.load_dataset("titanic")

# Set up a factorplot
g = sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", palette="muted", legend=False)
                   
# Show plot
plt.show()
```
